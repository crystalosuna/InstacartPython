# Create Busiest Days Column

```python
#get the frequency of 'orders_day_of_week'
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

The two busiest days are Saturday '0' and Sunday '1'. The two slowest days are Wednesday '4' and Tuesday '3'.

```python
#create a for-loop to get a result list for the busiest days based on values of 'orders_day_of_week'
days = []

for day in df_orders_products_merged['orders_day_of_week']:
  if day == 0 or day == 1:
    days.append('Busiest days')
  elif day == 4 or day == 3:
    days.append('Slowest days')
  else:
    days.append('Regular days')
```

```python
# create the 'busiest_days' column
df_orders_products_merged['Busiest_days'] = days
```

```python
df_orders_products_merged['Busiest_days'].value_counts (dropna = False)
```

```python
# check the output of the 'orders_day_of_week' and the 'busiest_days' column for accuracy
print('Output of the orders_day_of_week and the busiest_days column:\n', df_orders_products_merged[['orders_day_of_week', 'Busiest_days']].head(10))
```

# Create Busiest_period_of_day column

```python
#get the frequency of 'order_hour_of_day'
df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

Since there are 24 hours in total the number of hours can be devided into three groups. Hours with fewest orders, average orders, and most orders. Fewest orders(3, 4, 2, 5, 1, 0, 6, 23) Average orders (22, 21, 7, 20, 19, 18, 8, 17) Most oders(9, 16, 12, 13, 15, 14, 11, 10).

```python
#create a for-loop to get a result list for the busiest hours based on values of 'orders_day_of_week'
hours = []

for hour in df_orders_products_merged['order_hour_of_day']:
  if hour in [3, 4, 2, 5, 1, 0, 6, 23]:
    hours.append('Fewest orders')
  elif hour in [9, 16, 12, 13, 15, 14, 11, 10]:
    hours.append('Most orders')
  else:
    hours.append('Average orders')
```

```python
# create the 'Busiest_period_of_day' column
df_orders_products_merged['Busiest_period_of_day'] = hours
```

```python
# check the output of the 'orders_day_of_week' and the 'busiest_days' column for accuracy
df_orders_products_merged[['order_hour_of_day', 'Busiest_period_of_day']].head(10)
```

```python
# get the frequency of the values in the 'busiest_period_of_day' column
print('Frequency of values in the Busiest_period_of_day column:\n', df_orders_products_merged['Busiest_period_of_day'].value_counts())
```
